```python
with open('../pickles/test_train.dat', 'rb') as file:
    train_test_list = pickle.load(file)
    
X_train = train_test_list[0]
```

```python
char_dict = defaultdict(int)

def helper_cleaner(n_gram):
    # pound added in order to handle occurence of letter as either first or last in the string
    polish_letters_extended = 'aąbcćdeęfghijklłmnńoóprsśtuwyzźż' + "#" + 'qxv'
    non_polish_letters = set(n_gram) - set(polish_letters_extended)
    if non_polish_letters:
        return False
    return True

def generate_character_n_grams(text, n=2):
    
    global char_dict
    
    # handle first and last letter as in the example: słowo -> '#s', 'sł', ..., 'wo', 'o#'
    text = '#' + text + '#'
    
    all_grams = [text[i:i+n] for i in range(len(text)-1)]
    
    # n-grams contain non-Polish alphabet letters - delete such n-grams
    for n_gram in all_grams:
        if helper_cleaner(n_gram):
            # {'a' : {'a': 1, 'b': 3, ..., }, ... 'ż': {'a': 1, ...,}}
            try:
                char_dict[n_gram[0]][n_gram[1]] += 1
            except:
                print(n_gram)
#     clean_grams = [n_gram for n_gram in all_grams if helper_cleaner(n_gram)]
    
```

```python
char_dict = defaultdict(int)

for letter in 'aąbcćdeęfghijklłmnńoóprsśtuwyzźż' + '#' + 'qxv':
    char_dict[letter] = defaultdict(int)

for text_with_error in X_train['text_with_error'].values:
    generate_character_n_grams(text_with_error)
```

```python
with open('../pickles/char_dict.p', 'wb') as file:
    pickle.dump(char_dict, file, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
polish_letters_extended = 'aąbcćdeęfghijklłmnńoóprsśtuwyzźż' + "#" + 'qxv'

zero_data = np.zeros(shape=(len(polish_letters_extended), len(polish_letters_extended)))
```

```python
char_matrix = pd.DataFrame(copy.deepcopy(zero_data), columns=list(polish_letters_extended), index=list(polish_letters_extended))
char_matrix = char_matrix.astype(int)

for row_letter, sub_dict in char_dict.items():
    for col_letter, count in sub_dict.items():
        char_matrix[col_letter][row_letter] += count
```
